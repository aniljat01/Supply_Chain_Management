### Creating Output Data 1

```python
##Importing libraries
import numpy as np
import pandas as pd
```

```python
##Loading the data
x1 = pd.read_excel("Company X - Order Report.xlsx")
x2 = pd.read_excel("Company X - Pincode Zones.xlsx")
x3 = pd.read_excel("Company X - SKU Master.xlsx")
courier1 = pd.read_excel("Courier Company - Invoice.xlsx")
courier2 = pd.read_excel("Courier Company - Rates.xlsx")
```

```python
##Showing the top 5 rows of x1
x1.head()
```

```python
##Renamed the column to make it easier to merge the data
x1.rename(columns = {"ExternOrderNo" : "Order ID"}, inplace = True)
x1
```

```python
##Showing the top 5 rows of x2
x2.head()
```

```python
##Renamed the column to make it easier to understand in merged data
x2.rename(columns = {"Zone" : "Delivery Zone X"}, inplace = True)
x2.head()
```

```python
##Showing the top 5 rows of x3
x3.head()
```

```python
##Renamed the column to make it easier to understand in merged data
x3.rename(columns = {"Weight (g)" : "Total weight X (KG)"}, inplace = True)
##Convert the column from "gram" to "kilogram"
x3["Total weight X (KG)"] = x3["Total weight X (KG)"]/1000
x3.head()
```

```python
##Creating the weight slab for x comapny
x3["Weigt slab x (KG)"] = np.where(x3["Total weight X (KG)"]<0.501,0.5,
                                  np.where(x3["Total weight X (KG)"]<1.01,1,
                                    x3["Total weight X (KG)"]))
```

```python
##Checking the weight slab column
x3.head()
```

```python
##Renamed the column to make it easier to understand in merged data
courier1.rename(columns = {"Charged Weight" : "Total weight Courier (KG)"}, inplace = True)
courier1.rename(columns = {"Zone" : "Delivery Zone Courier"}, inplace = True)
courier1.rename(columns = {"Billing Amount (Rs.)" : "Charges Bill Courier"}, inplace = True)
##Showing the top 5 rows of courier1 after renaming column
courier1.head()
```

```python
##Creating the weight slab for courier comapny
courier1["Weight slab courier"] = np.where(courier1["Total weight Courier (KG)"] < 0.51, 0.5,
                                                     np.where(courier1["Total weight Courier (KG)"] < 1.01, 1,
                                                     np.where(courier1["Total weight Courier (KG)"] < 1.51, 1.5,
                                                     np.where(courier1["Total weight Courier (KG)"] < 2.01, 2,
                                                     np.where(courier1["Total weight Courier (KG)"] < 2.51, 2.5,
                                                     np.where(courier1["Total weight Courier (KG)"] < 3.01, 3,
                                                             (courier1["Total weight Courier (KG)"])))))))
```

```python
##Showing top 5 rows after creating weight slab
courier1.head()
```

```python
##Merging the data which have common column(X1 or X3)
merge_data1 = x1.merge(x3, how = "inner")
##Showing top 5 rows after merging the data
merge_data1.head()
```

```python
##Merging the data which have common column(X2 or Courier1)
merge_data2 = x2.merge(courier1, how = "inner")
##Showing top 5 rows after merging the data
merge_data2.head()
```

```python
##Merging the merge_data1 and merge_data2
merged_data = merge_data1.merge(merge_data2, how = "inner")
##Showing top 5 rows after final merging the data
merged_data.head()
```

```python
##Calculate the expected charge as per X comapny
fixed_price = {  ##These are fixed rates by courier company for forwad charges
    "fwd_a_fixed":29.5,
    "fwd_a_additional":23.6,
    "fwd_b_fixed":33,
    "fwd_b_additional":28.3,
    "fwd_c_fixed":40.1,
    "fwd_c_additional":38.9,
    "fwd_d_fixed":45.4,
    "fwd_d_additional":44.8,
    "fwd_e_fixed":56.6,
    "fwd_e_additional":55.5,
    "rto_a_fixed":13.6,  ##These are fixed rates by courier company for rto charges
    "rto_a_additional":23.6,
    "rto_b_fixed":20.5,
    "rto_b_additional":28.3,
    "rto_c_fixed":31.9,
    "rto_c_additional":38.9,
    "rto_d_fixed":41.3,
    "rto_d_additional":44.8,
    "rto_e_fixed":50.7,
    "rto_e_additional":55.5
}

fixed_slab = 0.5 ##"This is a fixed weight slab where a fixed rate is applied on the first 0.5 of the weight slab, 
                    ##and an additional rate is applied on the rest."
Expected_charge_X = [] ##Making a list of expected charge amount

for i,j,k in zip(merged_data["Delivery Zone X"], merged_data["Weigt slab x (KG)"], merged_data["Type of Shipment"]):
    a = j - fixed_slab ##Here, I subtracted the first weight slab of 0.5 so that we can get 
                       ##the remaining weight slab on which we will apply the additional charge
    if i == "a":
        base_charge = fixed_price["fwd_a_fixed"]+((a/0.5)*fixed_price["fwd_a_additional"]) ##Formula to calculate forwad charges
        
        if k == "Forward and RTO charges":
            base_charge = (base_charge + fixed_price["rto_a_fixed"]+((a/0.5)*fixed_price["rto_a_additional"]))
                                                                                           ##Formula to calculate Forwad and Rto charges
        Expected_charge_X.append(base_charge) ##Append the charges amonunt 
        
    elif i == "b":
        base_charge = fixed_price["fwd_b_fixed"]+((a/0.5)*fixed_price["fwd_b_additional"])
        
        if k == "Forward and RTO charges":
            base_charge = base_charge + fixed_price["rto_b_fixed"]+((a/0.5)*fixed_price["rto_b_additional"])
        Expected_charge_X.append(base_charge)
    elif i == "c":
        base_charge = fixed_price["fwd_c_fixed"]+((a/0.5)*fixed_price["fwd_c_additional"])
        
        if k == "Forward and RTO charges":
            base_charge = base_charge + fixed_price["rto_c_fixed"]+((a/0.5)*fixed_price["rto_c_additional"])
        Expected_charge_X.append(base_charge)
    elif i == "d":
        base_charge = fixed_price["fwd_d_fixed"]+((a/0.5)*fixed_price["fwd_d_additional"])
        
        if k == "Forward and RTO charges":
            base_charge = base_charge + fixed_price["rto_d_fixed"]+((a/0.5)*fixed_price["rto_d_additional"])
        Expected_charge_X.append(base_charge)
    elif i == "e":
        base_charge = fixed_price["fwd_e_fixed"]+((a/0.5)*fixed_price["fwd_e_additional"])
        
        if k == "Forward and RTO charges":
            base_charge = base_charge + fixed_price["rto_e_fixed"]+((a/0.5)*fixed_price["rto_e_additional"])
        Expected_charge_X.append(base_charge)
```

```python
print(Expected_charge_X)
```

```python
##Making a column in merged_data and fill all the values which are calculate earlier 
merged_data["Expected charge X"] = Expected_charge_X
```

```python
##Making a new column and Calculate Difference Between Expected Charges as per x and Billed Charged by courier company
merged_data["Difference expected charge and charges bill"] = merged_data["Charges Bill Courier"] - merged_data["Expected charge X"]
```

```python
##Dropping the columns that are not very necessary in the merged data
merged_data.drop("Warehouse Pincode", axis = 1, inplace = True)
merged_data.drop("Customer Pincode", axis = 1,inplace = True)
merged_data.drop("Order Qty", axis = 1,inplace = True)
```

```python
##Final data (Output 1)
new_order = ["Order ID", "AWB Code", "Total weight X (KG)", "Weigt slab x (KG)", "Total weight Courier (KG)", "Weight slab courier",
            "Delivery Zone X", "Delivery Zone Courier", "Expected charge X", "Charges Bill Courier", "Difference expected charge and charges bill"]
merged_data = merged_data[new_order]
merged_data.head()
```

```python
##How many delivery zone are similar of both comapnies
a = (merged_data["Delivery Zone X"] == merged_data["Delivery Zone Courier"]).sum()
a
```

```python
##How many weight slab are equal of both companies
b = (merged_data["Weigt slab x (KG)"] == merged_data["Weight slab courier"]).sum()
b
```

```python
##Converting Dataframe into CSV file
merged_data.to_csv("Output 1.csv", index = False)
```

```python
##Total Orders where X has been overcharged
b = (merged_data["Expected charge X"] < merged_data["Charges Bill Courier"]).sum()
b
```

```python
##Total Amount of Orders where X has been overcharged
condition = merged_data["Expected charge X"] < merged_data["Charges Bill Courier"]
total_sum = merged_data[condition]["Charges Bill Courier"].sum()
total_sum
```

```python
##Total Orders where X has been undercharged
c = (merged_data["Expected charge X"] > merged_data["Charges Bill Courier"]).sum()
c
```

```python
##Total Amount of Orders where X has been undercharged
condition = merged_data["Expected charge X"] > merged_data["Charges Bill Courier"]
total_sum = merged_data[condition]["Expected charge X"].sum()
total_sum
```
